## Update Course Formats

```python
re.sub("[^.,?A-Za-z0-9]+", ' ', df.loc[0]['full-time courses']).strip().replace(' , ', ', ')
```

```python
for ft in range(0, len(df)):
    df.loc[ft]['full-time courses'] = re.sub("[^,A-Za-z0-9]+", 
                                ' ', 
                                df.loc[ft]['full-time courses']).strip().replace(' , ', ', ')
```

```python
for pt in range(0, len(df)):
    df.loc[pt]['part-time courses'] = re.sub("[^,A-Za-z0-9]+", 
                                ' ', 
                                df.loc[ft]['part-time courses']).strip().replace(' , ', ', ')
```

```python
url = 'https://generalassemb.ly/locations/'
urls = []

for city in df['city']:
    if city == 'Tanjong Pagar':
        urls.append(url + 'singapore/' + city.replace(' ', '-').lower())
    else:
        urls.append(url + city.replace(' ', '-').lower())
```

```python
df['url'] = urls
```
